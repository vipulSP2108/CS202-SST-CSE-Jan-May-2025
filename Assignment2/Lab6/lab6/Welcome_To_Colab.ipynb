```
import matplotlib.pyplot as plt

# Data for Speedup Plot
parallel_modes = ['Sequential', '--dist load (1)', '--dist no (1)', '--dist load (auto)', '--dist no (auto)']
execution_times = [2.714, 2.97, 2.903, 26.174, 27.26]  # in seconds
speedups = [execution_times[0] / t for t in execution_times]

# Speedup Plot
plt.figure(figsize=(8, 5))
plt.bar(parallel_modes, speedups, color='lightblue')
plt.xlabel('Parallelization Mode')
plt.ylabel('Speedup')
plt.title('Speedup Comparison of Parallel Execution')
plt.show()
```
